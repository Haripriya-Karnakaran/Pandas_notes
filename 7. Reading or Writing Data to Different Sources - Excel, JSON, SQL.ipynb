```python
filt = (df["Country"] == "India")
india_df = df.loc[filt]            # here we making india as a separate df
```

```python
# to convert this india df to a csv file

india_df.to_csv("Dataset/India_csv.csv") # now we can see this csv in dataset folder
```

```python
# tsv file - Tab-separated values (TSV) is a simple, text-based file format for storing tabular data. 
# Records are separated by newlines, and values within a record are separated by tab characters. 
# The TSV format is thus a delimiter-separated values format, similar to comma-separated values

# In simple, tsv files are like converting a tsv from csv by using using tab as a delimiter instead of comma, which is used for making table easily
```

```python
# tsv

india_df.to_csv("Dataset/India_tsv.tsv",sep = "\t") # here we are spearating each data with tabs
```

```python
from sqlalchemy import create_engine  # engine to connect with database
import mysql
```

```python
engine = create_engine("mysql://root:3902@localhost/customer") # username-root and pass,database name
```

```python
# We are going to import our df as a table to sql database
# if_exists - "replace" is for if the table which we creating is already existed in the database means,it will replace it into new one
india_df.to_sql("sample_table", engine, schema=None, if_exists='replace', index=False)  # sample table is a name we want to be our df
```
