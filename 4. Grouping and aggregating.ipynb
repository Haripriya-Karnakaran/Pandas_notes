### grouping 
```python
             Grouping in pandas involves splitting the data into groups based on some criteria, 
             applying a function to each group independently, and then combining the results. 
             This is particularly useful for performing operations on subsets of data and aggregating results.
```

````python
df1["Country"].value_counts()
````

````python
country_grp = df1.groupby(["Country"])
````

````python
country_grp.get_group("United States of America") # here only the data which are united states has been generated
````

````python
country_grp.get_group("India")  # only india
````

````python
# we can get the same using filter

filt = df1["Country"] == "India"     # only india is here
df1.loc[filt]
````

````python
# to group something by another column
country_grp["RemoteWork"].value_counts()
````

````python
country_grp["RemoteWork"].value_counts().head(50)
````

````python
# to see only for specific object eg: only for india

country_grp["RemoteWork"].value_counts().loc["India"]
````

````python
country_grp["RemoteWork"].value_counts(normalize = True).loc["India"]
````

````python
country_grp["ConvertedCompYearly"].median() # can get the median for each country
````

````python
country_grp["ConvertedCompYearly"].median().loc["Germany"] # to get the median for specific country we use loc
````

````python
# to get the median,mean 

country_grp["ConvertedCompYearly"].agg(["median","mean"])
````

````python
# to get mean,median aggregations for specific country
country_grp["ConvertedCompYearly"].agg(["median","mean"]).loc["Germany"]
````

````python
# to see how many have worked with python

filt = df1["Country"] == "India"
df1.loc[filt]["LanguageHaveWorkedWith"].str.contains("Python")#it gives boolean values as true means know python else false
````

````python
# sum() will work with boolean values too since true means 1 and flase means 0

filt = df1["Country"] == "India"
df1.loc[filt]["LanguageHaveWorkedWith"].str.contains("Python").sum()
````

````python
country_grp["LanguageHaveWorkedWith"].str.contains("Python").sum() 
# this wont work if we want to know how many knows python for each country
````

````python
# instead we use apply()
country_grp["LanguageHaveWorkedWith"].apply(lambda x : x.str.contains("Python").sum())
````

````python
# Practise questions

# what % of people from each country knows python?

country_respondents = df1["Country"].value_counts()
country_respondents
````

````python
country_uses_python = country_grp["LanguageHaveWorkedWith"].apply(lambda x : x.str.contains("Python").sum())
country_uses_python
````

````python
python_df = pd.concat([country_respondents,country_uses_python],axis = "columns",sort = False)
python_df    # country - about 18647 people said they are from usa and languagehaveworkedwith is python 10062
````

````python
python_df.rename(columns = {"Country":"No_of_Respondents","LanguageHaveWorkedWith":"Knows_python"},inplace = True)
python_df
````

````python
# create a new column for percentage for python people

python_df["Percent_knows_python"] = (python_df["Knows_python"] / python_df["No_of_Respondents"])*100
python_df
````

````python
python_df.sort_values(by = "Percent_knows_python",ascending = False,inplace = True)
python_df
````
