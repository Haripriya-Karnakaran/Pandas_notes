```python
number = {"numbers" : [20,300,1,50,1000,35,32,76,83,23],"name" : ["Hari","priya","kumar","abi","ram","ashwini","zypher","sri","karuna","prema"]}
import pandas as pd
df = pd.DataFrame(number)
```

```python
df.sort_values(by = "numbers")
```

```python
df.sort_values(by = "numbers",ascending = False)
```

```python
# to sort with multiple columns

df.sort_values(by = ["numbers","name"],ascending = False)
```

```python
# to sort one column in ascending and one column in descending

df.sort_values(by = ["numbers","name"],ascending = [False,True],inplace = True) # false(descending) for numbers,true(ascending) for name
```

```python
df
```

```python
# now, to sort the data by index

df.sort_index()
```

```python
# to sort a single column as series

df["name"].sort_values()
```

```python
df["numbers"].sort_values()
```
