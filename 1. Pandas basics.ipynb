### Part 4 - Filters

```python
people = {"First_name":["Haripriya","xyz","Sathish"],
          "Last_name":["Karuna","abc","Karuna"],
          "age":[21,40,27],
          "email":["haripriya@gmail.com","sathish@gmail.com","xyz@gmail.com"]}
```

```python
import pandas as pd
df = pd.DataFrame(people)
df
```

```python
# to filter out only the last name which has "karuna"...

df["Last_name"] == "Karuna"
```

```python
# to get it as a dataframe or table, assign a variable
# we should not use "filter" as a variable name beacuse it is a method

filt = (df["Last_name"] == "Karuna")
```

```python
df[filt]
```

```python
# some may do the same thing like this but it could be confused sometimes

df[(df["Last_name"] == "Karuna")]    # we just directly giving it without using any variable
```

```python
# can use loc as well

df.loc[filt]
```

```python
df.loc[filt,"email"]    # to have only email column
```

```python
df.loc[filt,["age","email"]]
```

```python
# & - AND , | - OR

filt1 = (df["Last_name"] == "Karuna") & (df["First_name"] == "Haripriya")
```

```python
df[filt1]
```

```python
filt2 = (df["First_name"] == "Sathish") & (df["age"] == 27)
```

```python
df[filt2]
```

```python
# OR

filt3 = (df["First_name"] == "xyz") | (df["age"] == 21)
df[filt3]
```

```python
# to get the rest of the data

df.loc[~filt3]    # use ~(tilt)
```
