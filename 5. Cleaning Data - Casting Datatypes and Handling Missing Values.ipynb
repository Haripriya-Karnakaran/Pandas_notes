```python
import pandas as pd
import numpy as np
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}

# np.nan means not an number so np.nan,none everything means to be missing values
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.dropna() # by default, there is a approach here - df.dropna(axis = index, how = "any")
```

```python
df.dropna(axis = "index",how = "any") # same o/p like before...based on index, if any one is missing value in a row, that will be removed
```

```python
df.dropna(axis = "index",how = "all") # all is for if all the columns in a row has missing value then would be removed
```

```python
# cehcking which are considered as missing

check = {"a" : [np.nan,"NA",None,"missing","NA",None],"b":[np.nan,np.nan,np.nan,None,"NA",np.nan],
         "c":["missing","NA","missing","NA",None,None]}

df1 = pd.DataFrame(check)  # None, np.nan are the missing values 
```

```python
df1.dropna(axis = "index",how = "all") # here fifth row is missing since every data in the fifth row are missing values
```

```python
df1.dropna(axis = "index",how = "any") # none of the row is printed becuase every column has atleast one missing values
```

```python
# we can do the same based on column wise

df.dropna(axis = "columns",how = "all") # none column having missing values in all
```

```python
df.dropna(axis = "columns",how = "any") # it gives empty since every column has atleast one missing value
```

```python
# if we have none but we need a specific column or row even if we have missing values

df.dropna(axis = "index",how = "any",subset = ["email"]) # using subset
# we need email column so even any index has none it is printing email id email has missing values then it leaves it
```

```python
# now, if we want either this or that column if we have none , either this or that would be printed

df.dropna(axis = "index",how = "all",subset = ["last","email"])

# this is going to print like either last or email has none means it is going to print that column
```

```python
# to treat some missing values which we made like "NA" and "Missing"...to make this to  be considered 

import pandas as pd
import numpy as np
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}

df = pd.DataFrame(people)
```

```python
df
```

```python
# replace those with np.nan

df.replace("NA",np.nan,inplace = True)
df.replace("Missing",np.nan,inplace = True)
```

```python
df # everything is changed
```

```python
# now we can perform the same process like above

df.dropna()
```

```python
df.dropna(axis = "index",how = "any")
```

```python
df.dropna(axis = "index",how = "all")
```

```python
df.dropna(axis = "index",how = "all",subset = ["email"])
```

```python
# to check where are all the nan is present

df.isna()
```

```python
# to fill a value in all missing places

df.fillna("HARIPRIYA")
```

```python
df.fillna(0)
```
